```python

df['Salary'].fillna(df['Salary'].mean(), inplace=True)
```

```python
df['Salary'] = df['Salary'].fillna(df['Salary'].mean())
```

```python
df['Job_Satisfaction'] =df['Job_Satisfaction'].fillna(df['Job_Satisfaction'].mean())
```

```python

df['Work_Hours_Per_Week'] = df['Work_Hours_Per_Week'].fillna(df['Work_Hours_Per_Week'].mean())
```

```python
df
```

```python
df['Age'].fillna(df['Age'].mean(),inplace=True)
```

```python
print("Dataset after handling missing values:")
print(df)
```

```python

from sklearn.preprocessing import MinMaxScaler
```

```python

scaler = MinMaxScaler()
```

```python

df[['Age', 'Salary']] = scaler.fit_transform(df[['Age', 'Salary']])
```

```python
df['Job_Satisfaction'] = (df['Job_Satisfaction'] - df['Job_Satisfaction'].min()) / (df['Job_Satisfaction'].max() - df['Job_Satisfaction'].min())
```

```python

from sklearn.preprocessing import StandardScaler

s_scaler = StandardScaler()
```

```python

df[['Age']] = s_scaler.fit_transform(df[[ 'Age']])
```

```python
df['Age']
```

```python

df['Work_Hours_Per_Week'] = (df['Work_Hours_Per_Week'] - df['Work_Hours_Per_Week'].mean()) / df['Work_Hours_Per_Week'].std()
```
